```julia
using CairoMakie
using DataFrames

# Definimos dos funciones equivalentes
f(x) = sqrt(x^2+1) - 1
g(x) = x^2 / (sqrt(x^2+1)+1)

# Definimos un conjunto de valores de prueba
X = Float32[8.0^-n for n = 1:10];
```

```julia
# Construimos un dataframe para visualizar los datos
df = DataFrame("n" => 1:10, "f" => f.(X), "g" => g.(X))
show(df, row_labels=nothing)
```
