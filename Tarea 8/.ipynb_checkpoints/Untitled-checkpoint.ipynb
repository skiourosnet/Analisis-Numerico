# Error en el polinomio de interpolación para $\cos(x)$ y $e^x$

```python
from numpy import pi, e
from math import factorial
```

Función error de $f(x)=\cos(x)$ para $n+1$ nodos

```python
def error(n): # n = grado del polinomio
    err = pi ** (n+1) / (factorial(n+1) * 2 ** (n+1))
    return err

for n in range(1,101):
    err = round(error(n), ndigits=7) # Redondeamos el error a 7 posiciones decimales
    print(f"n = {n}  ", f"error = {err} ")

    if err < 10 ** -6: # Terminamos el ciclo al alcanzar la cota
        break
```

Función error de $f(x)=e^x$ para $n+1$ nodos

```python
def error(n): # n = grado del polinomio
    err = 2 ** (n+1) * e ** 2 / factorial(n+1) 
    return err

for n in range(1,101):
    err = round(error(n), ndigits=7) # Redondeamos el error a 7 posiciones decimales
    print(f"n = {n}  ", f"error = {err} ")

    if err < 10 ** -6: # Terminamos el ciclo al alcanzar la cota
        break
```
